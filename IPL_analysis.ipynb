## **Q12**

```python
def team_performance(df, team_name):
    df_1 = df[df['batting_team'] == team_name]
    df_2 = df[df['bowling_team'] == team_name]

    df_bat = df_1.groupby(['batting_team', 'bowling_team'])[['total_runs']].sum().sort_values(by='total_runs', ascending=False).reset_index()
    df_bat.columns = ['team_name', 'vs', 'total_runs']
    df_bowl = df_2.groupby(['bowling_team', 'batting_team'])[['player_dismissed']].count().reset_index()
    df_bowl.columns = ['team_name', 'vs', 'total_wickets']
    df_merged = pd.merge(df_bat, df_bowl, on=['team_name', 'vs'], how='inner')
    df = df_1.groupby(['match_id', 'bowling_team'])[['total_runs']].sum().reset_index()['bowling_team'].value_counts().reset_index()
    df_performance = pd.merge(df_merged, df, left_on='vs', right_on='bowling_team', how='inner')
    df_performance.rename(columns={'count': 'matches'}, inplace=True)
    df_performance.drop('bowling_team', axis=1, inplace=True)
    df_performance = df_performance.reindex(columns=['team_name', 'vs', 'matches', 'total_runs', 'total_wickets'])
    # for highest scores on each team
    df_scores = df_1.groupby(['match_id', 'bowling_team'])[['total_runs']].sum().reset_index()
    df_highest_scores = df_scores.groupby('bowling_team')[['total_runs']].max().reset_index()
    df_highest_scores.columns = ['vs', 'highest_score']
    df_performance = pd.merge(df_performance, df_highest_scores, on = 'vs', how='inner')

    return df_performance
```

```python
team_name = 'Royal Challengers Bangalore'
team_performance(df_ipl, team_name)
```

## **Q15**

```python
def player_performance(df, player_name):
    batting_data = {'name': player_name, 'team': np.nan, 'matches': np.nan, 'total_runs': np.nan, 'balls_faced': np.nan, 'avg_score': np.nan,
                       'highest_score': np.nan, 'strike_rate': np.nan, '100s': np.nan, '50s': np.nan, '4s': np.nan, '6s': np.nan}
    bowling_data = {'name': player_name, 'team': np.nan, 'matches': np.nan,
                       'wickets': np.nan, 'overs': np.nan, 'runs_given': np.nan, 'avg_runs_given': np.nan,
                       'Best_bowling_inning': np.nan, 'economy': np.nan, 'strike_rate': np.nan, '4w': np.nan, '5w': np.nan}
    df_batting_stats = pd.DataFrame(batting_data, columns=batting_data.keys(), index=[0])
    df_bowling_stats = pd.DataFrame(bowling_data, columns=bowling_data.keys(), index=[0])
    df_batting = df[df['batsman'] == player_name]
    df_bowling = df[df['bowler'] == player_name]
    # batting stats
    df_batting_stats['team'] = df_batting['batting_team'].iloc[0]
    df_batting_stats['matches'] = df_batting['match_id'].nunique()
    df_batting_stats['total_runs'] = df_ipl[df_ipl['batsman'] == player_name]['batsman_runs'].sum()
    df_batting_stats['balls_faced'] = df_batting['ball'].count()
    df_batting_stats['avg_score'] = (df_batting_stats['total_runs'] / df_batting_stats['matches']).round(2)
    df_batting_stats['highest_score'] = df_batting.groupby('match_id')['batsman_runs'].sum().max()
    df_batting_stats['strike_rate'] = ((df_batting_stats['total_runs'] / df_batting_stats['balls_faced']) * 100).round(2)

    df_runs = df_batting.groupby('match_id')['batsman_runs'].sum()   # batsman runs in each match
    df_batting_stats['100s'] = df_runs[df_runs >= 100].count()
    df_batting_stats['50s'] = df_runs[(df_runs >= 50) & (df_runs < 100)].count()
    df_batting_stats['4s'] = df_batting[df_batting['batsman_runs'] == 4]['batsman_runs'].count()
    df_batting_stats['6s'] = df_batting[df_batting['batsman_runs'] == 6]['batsman_runs'].count()
    # return df_batting_stats

    # bowling stats
    df_bowling_stats['team'] = df_bowling['bowling_team'].iloc[0]
    df_bowling_stats['matches'] = df_bowling['match_id'].nunique()
    # counts of wickets each category
    df_wickets = df_bowling[(df_bowling['dismissal_kind'] != 'obstructing the field') & (df_bowling['dismissal_kind'] != 'retired hurt') &
                    (df_bowling['dismissal_kind'] != 'run out')]
    df_bowling_stats['wickets'] = df_wickets.value_counts('dismissal_kind').sum()
    df_bowling_stats['overs'] = df_bowling['ball'].count() / 6
    df_bowling_stats['runs_given'] = (df_bowling['total_runs'] - df_bowling['penalty_runs']).sum()
    df_bowling_stats['avg_runs_given'] = (df_bowling_stats['runs_given'] / df_bowling_stats['matches']).round(2)
    # calculating wickets taken in each match
    df_wickets_per_m = df_wickets.groupby('match_id')['dismissal_kind'].count()
    most_wickets_id = df_wickets_per_m[df_wickets_per_m == df_wickets_per_m.max()].index[0]
    df_bowling_stats['Best_bowling_inning'] = str(df_bowling[df_bowling['match_id'] == most_wickets_id]['total_runs'].sum()) + '/' + str(df_wickets_per_m.max())
    df_bowling_stats['economy'] = (df_bowling_stats['runs_given'] / df_bowling_stats['overs']).round(2)
    df_bowling_stats['strike_rate'] = df_bowling['ball'].count()/df_bowling_stats['wickets']
    df_bowling_stats['4w'] = (df_wickets.groupby('match_id')['dismissal_kind'].count() == 4).sum()
    df_bowling_stats['5w'] = (df_wickets.groupby('match_id')['dismissal_kind'].count() == 5).sum()

    return df_batting_stats, df_bowling_stats
```
